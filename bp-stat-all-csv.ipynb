```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import ntpath

dfs = []
for filename in glob.glob("C:\\Users\\yury.shchetinin\\IdeaProjects\\battleplanes\\*.csv"):
    df = pd.read_csv(filename, names=['genome', 'result'])
    df["index"]= df.index
    dfs.append((df, ntpath.basename(filename)))
```

```python
colors = ['red', 'green','blue', 'orange', 'grey', 'yellow', 'black', 'brown', 'pink']
```

```python
%matplotlib inline
i = 0
fig, axes = plt.subplots(len(dfs), 1, figsize=(10, 5), sharex=True, sharey=True)
for df, filename in dfs:
    axes[i].plot(df['result'].apply(lambda x: x if x<= 1000 else x), color=colors[i], label=filename)
    axes[i].legend(loc='upper left')
    i+=1
plt.show()
```

```python
plt.figure(figsize=(8, 18))
i = 0
for df, filename in dfs:
    plt.boxplot(df['result'].apply(lambda x: x if x<= 1000 else 1001), positions=[i], labels=[filename])
    i+=1
```

```python
tdf = dfs[0][0]
```

```python
%matplotlib inline
i = 0
plt.grid
fig, axes = plt.subplots(len(dfs), 1, figsize=(10, 5), sharex=True, sharey=False)
for df, filename in dfs:
    gr = df.groupby('genome')
    axes[i].plot(gr['result'].quantile(0.9), color=colors[i], label=filename)
    axes[i].legend(loc='upper right')
    axes[i].grid()
    i+=1
plt.show()
```

```python
%matplotlib inline
i = 0
plt.scatter
fig, axes = plt.subplots(len(dfs), 1, figsize=(15, 15), sharex=True, sharey=True)
for df, filename in dfs:
    gr = df.groupby('genome').quantile(0.9)
    axes[i].scatter(gr.index, gr['result'], color=colors[i], label=filename)
    axes[i].legend(loc='upper right')
    i+=1
plt.show()
```
